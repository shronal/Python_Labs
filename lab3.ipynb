```python
import random

# Generate a random number between 1 and 9
number_to_guess = random.randint(1, 9)

while True:
    guess = int(input("Guess a number between 1 and 9: "))

    if guess == number_to_guess:
        print("Well guessed!")
        break
    else:
        print("Wrong guess. Try again!")
```

```python
odd_count=0
even_count=0
while True:
  n = int(input("Enter the numbers: "))
  if n % 2 == 0:
    even_count=even_count+1
  else:
    odd_count=odd_count+1
  if n == 0:
    break
print("Even count: ", even_count)
print("Odd count: ", odd_count)
```

```python
string_inp = input('Enter the string that consists of digits: ')
strkolen=0
alpkolen=0
for i in range (len(string_inp)):
  if string_inp[i].isalpha():
   strkolen=strkolen+1

  if string_inp[i].isdigit():
   alpkolen=alpkolen+1

print("Letters: ",strkolen)
print("Numbers: ",alpkolen)
```
