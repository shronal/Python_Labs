6. Write a Python program to get the key, value and item in a dictionary.

````python
data = {
    'name':'hari', 'age':'29', 'city':'Bhaktapur'
}

for key, value in data.items():
  print("Key:", key)
  print("Value:", value)
  print("Item:", (key, value))
  print("-------------------")
````

15. Write a Python program to sort a list alphabetically in a dictionary.
Sample: num_dict = {'n1': [2, 3, 1], 'n2': [5, 1, 2], 'n3': [3, 2, 4]}
Output: {'n1': [1, 2, 3], 'n2': [1, 2, 5], 'n3': [2, 3, 4]}

````python
num_dict = {'n1': [2, 3, 1], 'n2': [5, 1, 2], 'n3': [3, 2, 4]}
for x in num_dict.values():
  x.sort()
print(num_dict)
````

Write a Python program to create a dictionary from a string.
Note: Track the count of the letters from the string.
Sample string : ‘engineerings’

````python
strs = 'engineerings'
dict = {}
for x in strs:
  if x in dict:
    dict[x] += 1
  else:
    dict[x] = 1
print(dict)
````

This code snippet counts the occurrences of each character in the string "engineerings" and stores them in a dictionary.

First, a string named strs is assigned the value "engineerings". An empty dictionary named dict is created to store the character counts.

The code then iterates over each character in the string strs using a for loop. For each character x, it checks if the character is already a key in the dictionary dict.

If it is, the code increments the value associated with that key (character) by 1, indicating another occurrence of that character.

If the character is not already a key in the dictionary, the code adds the character as a key to the dictionary and sets its value to 1, indicating the first occurrence of that character.

Finally, the code prints the dictionary, which now contains the counts of each character in the string "engineerings".

18. Write a Python program to find the highest 3 values of corresponding keys in a dictionary.

````python
dict = {
    'n1' : 3 , 'n2' : 45 , 'n3' : 5, 'n4' : 56, 'n5' : 4
}
sorted = list(dict.values())
sorted.sort(reverse=True)
print("The highest top 3 is:")
sorted = sorted[:3]
print(sorted)
````

19. Write a Python program to drop empty items from a given dictionary.
Original Dictionary: {'c1': 'Red', 'c2': 'Green', 'c3': None}
New Dictionary after dropping empty items: {'c1': 'Red', 'c2': 'Green'}

````python
dict1 = {'c1': 'Red', 'c2': 'Green', 'c3': None}
if None in dict1.values():
  for key, value in list(dict1.items()):
    if value is None:
      del dict1[key]
print(dict1)
````
